```python
import numpy as np
import pandas as pd
import glob
```

```python
df_adhi = pd.read_csv("D:\mesindex\ADHI.csv", parse_dates=["Date"])
df_tlkm = pd.read_csv("D:\mesindex\TLKM.csv", parse_dates=["Date"])
df_pgas = pd.read_csv("D:\mesindex\PGAS.csv", parse_dates=["Date"])
df_bris = pd.read_csv("D:\mesindex\BRIS.csv", parse_dates=["Date"])
```

```python
all_data = pd.concat([df_adhi, df_tlkm, df_pgas, df_bris])
all_data.head()
```

```python
mesidx = all_data["Name"].unique()
mesidx
```
